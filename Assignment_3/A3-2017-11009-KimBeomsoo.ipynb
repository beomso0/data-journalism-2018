### Q5: 질문을 스스로 만들고 답을 제시하시오.

```python
#객실 분류별(A.B.C...) 가격과 생존율을 알아보자

#1 객실 표시가 된 데이터들만을 추출한다
cabin_data = data[pd.notnull(data.cabin)]

#2 cabin_data의 cabin 값들을 살펴본다

cabin_data.index = range(len(cabin_data))
cabin_data.cabin.unique()
```

```python
#3 'F G63', 'F E57', 'F E46', 'F G73', 'F E69' 는 한 사람에 두 가지의 알파벳이 할당되어 있고, 이에 대한 설명이 제공되어 있지 않으므로 제외한다

not_subject = ['F G63', 'F E57', 'F E46', 'F G73', 'F E69']

def refine(thelist, subject_data):
    crash=[]
    for idx,row in subject_data.iterrows():
        if row['cabin'] in thelist:
            crash.append(True)
        else:
            crash.append(False)
    subject_data['is_crash'] = crash
    return subject_data

refine(not_subject, cabin_data)
cabin_data = cabin_data[cabin_data.is_crash == False]
cabin_data.cabin.unique()
```

```python
#4 객실에 번호가 있으므로 알파벳으로만 분류한 column을 cabin_data에 추가한다
alphabets = ['A','B','C','D','E','F','G','T']
def alphabet(thelist, subject_data):
    simple=[]
    for idx,row in subject_data.iterrows():
        for al in alphabets:
            if al in row['cabin']:
                simple.append(al)
    subject_data['cabin_alphabet'] =simple
    return subject_data
alphabet(alphabets, cabin_data)
cabin_data.cabin_alphabet.describe()
```

```python
#5 객실 분류별 fare 평균과 생존율을 알아보자
for al,grouped_data in cabin_data.groupby('cabin_alphabet'):
    print('cabin category : {}   mean_fare : {}     survival_rate: {}'.format(al,round(grouped_data.fare.mean(),2),\
                                                                              round(grouped_data['survived'].sum()/len(grouped_data)*100, 2)))
```
