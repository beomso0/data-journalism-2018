#1. 다음을 출력하는 프로그램을 작성하시오.

> **100** little **monkeys** jumping on the bed.

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, "No more monkeys jumping on the bed!"

> ...

> ... 

> **1** little **monkey** jumping on the bed. 

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, 

> "No more monkeys jumping on the bed!" 

(참고) Mama 혹은 Daddy를 랜돔하게 선택하게 하자.

```python
from random import choice

#틀이 되는 문장 오브젝트 생성
a = """%d little monkey%s jumping on the bed. \n 
One fell off and broke %s head. \n 
%s called the doctor and the doctor said, \"No more monkeys jumping on the bed!\""""

# 1~100까지의 숫자 역순으로 이뤄진 리스트 생성
numbers = [x for x in range(1,101)]
numbers.reverse()

#랜덤하게 선택할 리스트 생성
hisorher = ["his","her"]
momordad = ["Mama", "Daddy"]

for x in numbers:
    if x!=1:
        print ((a % (x, "s", choice(hisorher), choice(momordad)))+"\n\n")
    else :
        print ((a % (x, "", choice(hisorher), choice(momordad)))+"\n\n")
    
```

#2. 시작하는 연도와 끝나는 연도를 물어보는 프로그램을 작성 하고, 그 사이에 있는 모든 윤년을 출력하라. 

(시작하는 연도와 끝나는 연도가 윤년이면 그 해도 출력) 

* 윤년은 4로 나누어짐 (예: 1984, 2012)
* 100으로 나누어지면 윤년이 아님 (예: 1800, 1900) 
* 윤년은 400으로 나누어짐 (예: 1600, 2000) 

> Pick a starting year (like 1973 or something): 1973

> Now pick an ending year: 1977 

> Check it out... these years are leap years: 1976 

(참고) 윤년 리스트: http://mwultong.blogspot.com/2005/12/18002300-leap-year-list.html

```python
# 여기에 코드를 입력하세요.

start = input("Pick a starting year  ")
end = input("Now pick an ending year  ")

intsta = int(start)
intend = int(end)

leaps = []

while intsta < intend:
    if intsta%4 == 0 and (intsta%100 != 0 or intsta%400 ==0):
        leaps.append(intsta)
    intsta += 1

print ("Check it out... these years are leap years:  " + str(leaps))
```

#3. 구구단을 출력하는 프로그램을 작성하고 결과를 txt 파일로 저장하시오.

```python
# 구구단 리스트 생성 
result = []
for x in range(2,10):
    for y in range(1,10):
        result.append("%d X %d = %d"%(x,y,x*y))

#파일에 구구단 저장
with open("text.txt","w") as my_text:
    for i in range(0,len(result)):
        my_text.write(str(result[i])+"\n")
    
```
